```python
# Data Visualization

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
x=np.arange(0,10)
y=np.arange(10,20)
```

```python
# Scatter Plot the graph
plt.scatter(x,y,c='g')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Scatter Plot')
plt.savefig('Image.png')
```

```python
# Plot graph

y=x*x

plt.plot(x,y,'r*',linewidth=2, linestyle='dashed',markersize=12)
```

```python
## Subplots: subplot(row,col,pos)

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
```

```python
# plot sin and cos 
x=np.arange(0,5*np.pi,0.1)

y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'b--')
```

```python
x=[1,2,3,4,5,6]
y=[1,4,9,16,25,30]

plt.bar(x,y,width=0.7)
plt.title('Bar Graph')
```

```python
x=[10,20,10,40,50,60,70,50,80,90]

plt.subplot(2,1,1)
plt.hist(x,bins=10)
plt.title('Histogram')

plt.subplot(2,1,2)
plt.hist(x,bins=20)
plt.title('Histogram')
```
